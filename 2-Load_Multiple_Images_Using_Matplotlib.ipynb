```python
import cv2
import matplotlib.pyplot as plt
```

```python
image1 = cv2.imread('../images/neal.png')
image2 = cv2.imread('../images/bahubali.png')
image3 = cv2.imread('../images/jacksparrow.png')
image4 = cv2.imread('../images/ironman.png')
```

```python
ax1 = plt.subplot2grid((10,10),(0,0),rowspan=5,colspan=5)
ax2 = plt.subplot2grid((10,10),(0,5),rowspan=5,colspan=5)
ax3 = plt.subplot2grid((10,10),(5,0),rowspan=5,colspan=5)
ax4 = plt.subplot2grid((10,10),(5,5),rowspan=5,colspan=5)
```

```python
ax1.imshow(image1)
ax2.imshow(image2)
ax3.imshow(image3)
ax4.imshow(image4)
plt.show()
```
